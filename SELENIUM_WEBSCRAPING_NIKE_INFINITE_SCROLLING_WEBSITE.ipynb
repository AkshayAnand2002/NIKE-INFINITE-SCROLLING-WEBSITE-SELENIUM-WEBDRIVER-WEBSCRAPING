```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
driver=webdriver.Chrome('C:/Users/aksha/OneDrive/Desktop/chromedriver.exe')
driver.get('https://www.nike.com/ca/w/')
```

```python
last_height=driver.execute_script('return document.body.scrollHeight')
```

```python
last_height
```

```python
while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    time.sleep(2)
    new_height=driver.execute_script('return document.body.scrollHeight')
    if new_height==last_height:
        break
    last_height=new_height;
```

```python
#above lines of code help in scrolling to the bottom of the page until no more
#products are available.
```

```python
soup=BeautifulSoup(driver.page_source,'lxml')
```

```python
product_card=soup.find_all('div',class_ ='product-card__body')
```

```python
len(product_card)
```

```python
df=pd.DataFrame({'LINK':[''],'NAME':[''],'SUBTITLE':[''],'PRICE':['']})
for product in product_card:
    try:
        link=product.find('a',class_='product-card__link-overlay').get('href')
        name=product.find('div',class_ ='product-card__title').text
        subtitle=product.find('div',class_='product-card__subtitle').text
        price=product.find('div',class_='product-price ca__styling is--current-price css-11s12ax').text[1:]
        df=df.append({'LINK':link,'NAME':name,'SUBTITLE':subtitle,'PRICE':price},ignore_index=True)
    except:
        pass
```
